                                        Machine Learning with Unsupervised Dataset

UNSUPERVISED LEARNING - It is one of the category in machine learning, where the dataset will be there without any labels. So we need to cluster each samples in the dataset based on their features. There is no correct answer unlike the Superviesd Learning. It is widely used in recommendation system and used by Netflix for movie recommendations, etc.

Importing Neccessary Packages and reading the csv file and printing the head of the csv file.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

file = pd.read_csv("quake.csv")
file.head()
```

Applying the Principal Component Analysis (PCA) to the dataset to reduce the 4 features to 2 features and applying the KMeans algorith with 5 clusters and plotting them in the scatter plot with cenroids.

One of the Evaluation metric - Silhouette_Score :

The Silhouette Coefficient is calculated using the mean intra-cluster distance (a) and the mean nearest-cluster distance (b) for each sample. The Silhouette Coefficient for a sample is (b - a) / max(a, b). To clarify, b is the distance between a sample and the nearest cluster that the sample is not a part of. Note that Silhouette Coefficient is only defined if number of labels is 2 <= n_labels <= n_samples - 1.

The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more similar.

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn import metrics

pca= PCA(n_components=2).fit(file)
pca_2d = pca.transform(file)

clustering=KMeans(n_clusters=5)
clustering.fit(pca_2d)
labels = clustering.labels_
centroids = clustering.cluster_centers_
print(labels)
print(centroids)

colors = ['g.','r.','b.','k.','m.']

for i in range(len(pca_2d)):
    print("Co-ordinates:",pca_2d[i],"Labels:",labels[i])
    plt.plot(pca_2d[i][0],pca_2d[i][1],colors[labels[i]],markersize=10)
plt.scatter(centroids[:,0],centroids[:,1],s=150,marker="x",linewidths=5,zorder=10)

print(metrics.silhouette_score(pca_2d,labels,metric='euclidean'))


```
