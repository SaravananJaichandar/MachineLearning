```python
from sklearn.cross_validation import train_test_split 
train,test = train_test_split(f,test_size=0.2)
print(train.shape)
print(test.shape)
features_train = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
features_test = test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
labels_train = train.Species
labels_test = test.Species
print(labels_test.head())
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
training_set= clf.fit(features_train,labels_train)
prediction = clf.predict([[5.1,3.8,1.5,0.3]])
print(prediction)
print("Accuracy:",clf.score(features_test,labels_test))
```

```python
from sklearn import svm
clf = svm.SVC(kernel="rbf")
training_set= clf.fit(features_train,labels_train)
prediction = clf.predict([[5.1,3.8,1.5,0.3]])
print(prediction)
print("Accuracy:",clf.score(features_test,labels_test))
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
training_set= clf.fit(features_train,labels_train)
prediction = clf.predict(features_test)
print(prediction)
print("Accuracy:",clf.score(features_test,labels_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
training_set= clf.fit(features_train,labels_train)
prediction = clf.predict(features_test)
print(prediction)
print("Accuracy:",clf.score(features_test,labels_test))
```
